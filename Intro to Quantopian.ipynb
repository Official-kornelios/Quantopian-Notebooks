```python
#import research functions
from quantopian.research import prices, symbols
```

```python
#import pandas
import pandas as pd
```

```python
#query historical pricing for AMAZON stocks
aapl_close = prices(
    assets = symbols('AMZN'),
    start = '2013-01-01',
    end = '2016-01-01',
)

#compute long (50-days) and short(20-days) moving averages on
#AMZN data
amzn_sma20 = amzn_close.rolling(20).mean()
amzn_sma50 = amzn_close.rolling(50).mean()

#plot data
pd.DataFrame({
    'AMZN': amzn_close,
    'SMA20': amzn_sma20,
    'SMA50': amzn_sma50,
}).plot(
    title = 'Amazon close price/ SMA crossover'
);
```
